```python
import os
import time
import numpy as np
import torch
import pandas as pd
import logging
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from astropy.io import fits
from PIL import Image
from torch import nn
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms
import timm
from transformers import AutoProcessor, AutoModelForImageClassification

# ============================ Set Logging ============================
logging.basicConfig(level=logging.INFO, format='%(asctime)s %(levelname)s:%(message)s')
logger = logging.getLogger(__name__)

# ============================ Define Data Path ============================
common_test_sample_path = '/home/parlange/datasets/AnupreetaMore/common_test_sample_HSC/'
# L2: pure lenses test set
L2 = os.path.join(common_test_sample_path, 'new_cand_L2/new_cand_test/')

# ============================ Define Colors for Models ============================
colors = {
    'ViT': 'crimson',
    'MLP-Mixer': 'tomato',
    'CvT': 'darkred',
    'Swin': 'indigo',
    'CaiT': 'royalblue',
    'DeiT': 'cadetblue',
    'DeiT3': 'dodgerblue',
    'Twins_SVT': 'lightgreen',
    'Twins_PCPVT': 'mediumseagreen',
    'PiT': 'dimgrey',
    'ResNet-18': 'gold',
    'Ensemble': 'darkgoldenrod',
    'Random': 'black'
}

# ============================ Utility Functions ============================
def normalize_image(image):
    min_val = np.nanmin(image)
    max_val = np.nanmax(image)
    if max_val - min_val == 0:
        return image
    return (image - min_val) / (max_val - min_val)

# Helper function to load and process a FITS image into an 8-bit RGB image for visualization
def load_and_process_fits_image(file_path):
    try:
        with fits.open(file_path) as hdul:
            data = hdul[0].data
            if data is None:
                raise ValueError(f"No data in FITS file: {file_path}")
    except Exception as e:
        logger.error(f"Error loading FITS image: {file_path}: {e}")
        raise e
    if len(data.shape) == 2:
        data = np.stack((data,) * 3, axis=-1)
    elif len(data.shape) == 3:
        if data.shape[0] == 3:
            data = np.transpose(data, (1, 2, 0))
    rgb_image = np.stack([
        normalize_image(data[:, :, 2]),  # i-band → Red
        normalize_image(data[:, :, 1]),  # r-band → Green
        normalize_image(data[:, :, 0])   # g-band → Blue
    ], axis=-1)
    image_uint8 = (rgb_image * 255).astype(np.uint8)
    return image_uint8

# ============================ Define FITSImageDataset ============================
class FITSImageDataset(Dataset):
    def __init__(self, file_paths, labels, transform=None, normalize=True):
        if len(file_paths) != len(labels):
            raise ValueError("File paths and labels must have the same length.")
        self.file_paths = file_paths
        self.labels = labels
        self.transform = transform
        self.normalize = normalize
        logger.info(f"Total samples: {len(self.file_paths)}")

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        file_path = self.file_paths[idx]
        label = self.labels[idx]
        try:
            with fits.open(file_path) as hdul:
                data = hdul[0].data
                if data is None:
                    raise ValueError(f"No data found in FITS file: {file_path}")
        except Exception as e:
            logger.error(f"Error loading FITS file {file_path}: {e}")
            raise RuntimeError(f"Failed to load file: {file_path}") from e

        if len(data.shape) == 2:
            data = np.stack((data,) * 3, axis=-1)
        elif len(data.shape) == 3:
            if data.shape[0] == 3:  # (C, H, W) → (H, W, C)
                data = np.transpose(data, (1, 2, 0))
            elif data.shape[2] != 3:
                raise ValueError(f"Unexpected image shape {data.shape} for file {file_path}.")
        else:
            raise ValueError(f"Invalid image dimensions {data.shape} for file {file_path}.")

        try:
            rgb_image = np.stack([
                normalize_image(data[:, :, 2]),  # i-band → Red
                normalize_image(data[:, :, 1]),  # r-band → Green
                normalize_image(data[:, :, 0])   # g-band → Blue
            ], axis=-1)
        except IndexError as e:
            logger.error(f"Error mapping bands for file: {file_path}.")
            raise RuntimeError(f"Failed to map bands for file: {file_path}") from e

        if self.transform:
            image = Image.fromarray((rgb_image * 255).astype(np.uint8))
            image = self.transform(image)
        else:
            image = torch.from_numpy(rgb_image.transpose((2, 0, 1))).float()

        label = torch.tensor(label, dtype=torch.long)
        return image, label

def check_fits_files(file_paths, labels):
    valid_files = []
    valid_labels = []
    for idx, file_path in enumerate(file_paths):
        try:
            with fits.open(file_path, memmap=False) as hdul:
                hdul.verify('silentfix')
            valid_files.append(file_path)
            valid_labels.append(labels[idx])
        except Exception as e:
            logger.warning(f"Corrupted file detected: {file_path}, Error: {e}")
    return valid_files, valid_labels

def load_files_and_labels(directory, label):
    try:
        files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.fits')]
        labels = [label] * len(files)
        return files, labels
    except Exception as e:
        logger.error(f"Error accessing directory {directory}: {e}")
        return [], []

# ============================ Model Initialization ============================
def initialize_model(model_name, num_classes, pretrained=True, stochastic_depth_prob=0.1, 
                     fine_tune_mode="last_n_blocks", last_n=6):
    def freeze_all(model):
        for p in model.parameters():
            p.requires_grad = False

    def unfreeze_head(model):
        if hasattr(model, 'head') and isinstance(model.head, nn.Module):
            for p in model.head.parameters():
                p.requires_grad = True
        elif hasattr(model, 'classifier') and isinstance(model.classifier, nn.Module):
            for p in model.classifier.parameters():
                p.requires_grad = True
        else:
            for name, module in model.named_modules():
                if 'head' in name or 'classifier' in name:
                    for p in module.parameters():
                        p.requires_grad = True

    def unfreeze_last_n_blocks(blocks, n):
        if n > len(blocks):
            n = len(blocks)
        for block in blocks[-n:]:
            for p in block.parameters():
                p.requires_grad = True

    def unfreeze_all_blocks(blocks):
        for block in blocks:
            for p in block.parameters():
                p.requires_grad = True

    if model_name == 'ViT':
        model = timm.create_model('vit_base_patch16_224.augreg2_in21k_ft_in1k',
                                  pretrained=pretrained, drop_path_rate=stochastic_depth_prob)
        model.head = nn.Linear(model.head.in_features, num_classes)
        ft_attr = 'blocks'
    elif model_name == 'CvT':
        cvt_model_name = "microsoft/cvt-21"
        _ = AutoProcessor.from_pretrained(cvt_model_name)
        model = AutoModelForImageClassification.from_pretrained(
            cvt_model_name,
            num_labels=num_classes,
            ignore_mismatched_sizes=True
        )
        ft_attr = None
    elif model_name == 'CaiT':
        model = timm.create_model('cait_s24_224.fb_dist_in1k',
                                  pretrained=pretrained, drop_path_rate=stochastic_depth_prob)
        model.head = nn.Linear(model.head.in_features, num_classes)
        ft_attr = 'blocks'
    elif model_name == 'DeiT':
        model = timm.create_model('deit_base_patch16_224.fb_in1k',
                                  pretrained=pretrained, drop_path_rate=stochastic_depth_prob)
        model.head = nn.Linear(model.head.in_features, num_classes)
        ft_attr = 'blocks'
    elif model_name == 'DeiT3':
        model = timm.create_model('deit3_base_patch16_224.fb_in1k',
                                  pretrained=pretrained, drop_path_rate=stochastic_depth_prob)
        model.head = nn.Linear(model.head.in_features, num_classes)
        ft_attr = 'blocks'
    elif model_name == 'Swin':
        model = timm.create_model('swin_base_patch4_window7_224',
                                  pretrained=pretrained, drop_path_rate=stochastic_depth_prob,
                                  num_classes=num_classes)
        ft_attr = 'layers'
    elif model_name == 'Twins_SVT':
        model = timm.create_model('twins_svt_base.in1k',
                                  pretrained=pretrained, drop_path_rate=stochastic_depth_prob)
        model.head = nn.Linear(model.head.in_features, num_classes)
        ft_attr = 'stages' if hasattr(model, 'stages') else None
    elif model_name == 'Twins_PCPVT':
        model = timm.create_model('twins_pcpvt_base.in1k',
                                  pretrained=pretrained, drop_path_rate=stochastic_depth_prob)
        model.head = nn.Linear(model.head.in_features, num_classes)
        ft_attr = 'stages' if hasattr(model, 'stages') else None
    elif model_name == 'PiT':
        model = timm.create_model('pit_b_224.in1k',
                                  pretrained=pretrained, drop_path_rate=stochastic_depth_prob)
        model.head = nn.Linear(model.head.in_features, num_classes)
        ft_attr = 'blocks'
    elif model_name == 'MLP-Mixer':
        model = timm.create_model('mixer_b16_224.goog_in21k_ft_in1k',
                                  pretrained=pretrained, drop_path_rate=stochastic_depth_prob)
        model.head = nn.Linear(model.head.in_features, num_classes)
        ft_attr = 'blocks'
    elif model_name == 'ResNet-18':
        model = timm.create_model('resnet18',
                                  pretrained=pretrained,
                                  num_classes=num_classes)
        model.head = nn.Linear(model.fc.in_features, num_classes)
        model.fc = model.head
        model.layer_blocks = [model.layer1, model.layer2, model.layer3, model.layer4]
        ft_attr = 'layer_blocks'
    else:
        raise ValueError(f"Model {model_name} not recognized.")

    # Freeze everything initially
    freeze_all(model)
    # Unfreeze only the head for classification
    unfreeze_head(model)

    # Depending on the fine-tuning mode:
    if fine_tune_mode == "classification_head":
        return model
    if fine_tune_mode == "all_blocks":
        if ft_attr and hasattr(model, ft_attr):
            blocks = getattr(model, ft_attr)
            unfreeze_all_blocks(blocks)
        else:
            # Fallback: unfreeze everything
            for n, p in model.named_parameters():
                p.requires_grad = True
    elif fine_tune_mode == "last_n_blocks":
        if ft_attr and hasattr(model, ft_attr):
            blocks = getattr(model, ft_attr)
            unfreeze_last_n_blocks(blocks, last_n)
        else:
            # Fallback if the model doesn't have "blocks"/"layers" or similar
            all_params = list(model.named_parameters())
            frozen = [(n, p) for (n, p) in all_params if not p.requires_grad]
            if last_n > len(frozen):
                last_n = len(frozen)
            for (n, p) in frozen[-last_n:]:
                p.requires_grad = True
    else:
        raise ValueError("fine_tune_mode must be one of: 'classification_head', 'last_n_blocks', 'all_blocks'")
    return model

# ============================ Evaluation Function ============================
def evaluate_on_L2(models, transform, num_classes, device, experiment_folder):
    """
    Evaluate each model in `models` on L2.
    This function tracks the false negatives (i.e. lens samples that were missed, where the prediction is 0 even though ground truth is 1)
    for each model.
    Returns:
        - results: a dictionary with recall, detection count, false negative count and total samples per model.
        - false_negatives: a dictionary with the list of file paths (for each model) that were false negatives.
        - file_paths: list of all file paths in the L2 dataset.
    """
    results = {}
    false_negatives = {}  # To store missed lens file paths per model
    overall_start_time = time.time()

    file_paths, labels = load_files_and_labels(L2, label=1)
    file_paths, labels = check_fits_files(file_paths, labels)
    test_dataset = FITSImageDataset(file_paths, labels, transform=transform)
    test_loader = DataLoader(test_dataset, batch_size=256, shuffle=False, num_workers=2,
                             pin_memory=True if device.type == 'cuda' else False)

    for model_name, model in models.items():
        start_time = time.time()
        logger.info(f'\n=== Evaluating {model_name} on L2 ===')
        false_negatives[model_name] = []  # Initialize list for false negatives

        model.eval()
        model = model.to(device)

        # Load model weights from the experiment folder
        model_save_path = os.path.join(experiment_folder, f'{model_name}_best.pth')
        if os.path.exists(model_save_path):
            try:
                state_dict = torch.load(model_save_path, map_location=device)
                # Handle DataParallel or standard keys
                if list(state_dict.keys())[0].startswith("module."):
                    from collections import OrderedDict
                    new_state_dict = OrderedDict()
                    for k, v in state_dict.items():
                        new_state_dict[k[7:]] = v
                    model.load_state_dict(new_state_dict)
                else:
                    model.load_state_dict(state_dict)
                logger.info(f"Loaded weights for {model_name} from {model_save_path}")
            except Exception as e:
                logger.error(f"Error loading weights for {model_name}: {e}")
                continue
        else:
            logger.warning(f"No saved weights for {model_name} in '{experiment_folder}'. Skipping evaluation.")
            continue

        all_labels, all_preds = [], []
        detection_count = 0
        false_negative_count = 0
        current_index = 0

        with torch.no_grad():
            for images, batch_labels in tqdm(test_loader, desc=f'Evaluating {model_name} on L2'):
                images = images.to(device, non_blocking=True)
                batch_labels = batch_labels.to(device, non_blocking=True)
                outputs = model(images)
                logits = outputs.logits if hasattr(outputs, 'logits') else outputs
                _, predicted = torch.max(logits.data, 1)
                
                batch_labels_np = batch_labels.cpu().numpy()
                preds_np = predicted.cpu().numpy()
                all_labels.extend(batch_labels_np)
                all_preds.extend(preds_np)
                detection_count += (preds_np == 1).sum()
                batch_size = images.size(0)
                
                # Record false negatives: ground truth is 1 but prediction is 0.
                for i in range(batch_size):
                    if batch_labels_np[i] == 1 and preds_np[i] == 0:
                        false_negatives[model_name].append(test_dataset.file_paths[current_index + i])
                        false_negative_count += 1
                current_index += batch_size

        recall = (np.array(all_preds) == np.array(all_labels)).mean()
        results[model_name] = {
            'Recall': recall,
            'Detections': detection_count,
            'FalseNegatives': false_negative_count,
            'Total': len(test_dataset)
        }
        logger.info(f"{model_name} on L2 - Recall: {recall*100:.2f}%, Detections: {detection_count}/{len(test_dataset)}, False Negatives: {false_negative_count}")
        model_eval_time = time.time() - start_time
        logger.info(f"Evaluation time for {model_name}: {model_eval_time:.2f} seconds.")

    total_eval_time = time.time() - overall_start_time
    logger.info(f"Overall evaluation on L2 for experiment '{experiment_folder}' completed in {total_eval_time:.2f} seconds.")
    return results, false_negatives, file_paths

# ============================ Main Execution for All Experiments ============================
if __name__ == '__main__':
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    logger.info(f'Using device: {device}')

    # Define a standard transform
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225])
    ])

    # Models to evaluate
    model_names = [
        'ViT', 'MLP-Mixer', 'CvT', 'Swin', 'CaiT', 
        'DeiT', 'DeiT3', 'Twins_SVT', 'Twins_PCPVT', 'PiT', 'ResNet-18'
    ]

    # Paths to each experiment folder
    experiment_folders = [
        '/home/parlange/models/gravit/a1-C21-classification-head',
        '/home/parlange/models/gravit/a2-C21-half',
        '/home/parlange/models/gravit/a3-C21-all-blocks-ResNet18',
        '/home/parlange/models/gravit/b1-J23-classification-head',
        '/home/parlange/models/gravit/b2-J23-half',
        '/home/parlange/models/gravit/b3-J23-all-blocks',
        '/home/parlange/models/gravit/c1-C21+J23-classification-head',
        '/home/parlange/models/gravit/c2-C21+J23-half',
        '/home/parlange/models/gravit/c3-C21+J23-all-blocks-ResNet18',
        '/home/parlange/models/gravit/s1-C21-classification-head-18660',
        '/home/parlange/models/gravit/s2-C21-half-18660',
        '/home/parlange/models/gravit/s3-C21-all-blocks-ResNet18-18660'
    ]

    # Number of classes in your classification problem (lens vs non-lens)
    num_classes = 2

    # We'll collect per-experiment summary DataFrames in order to compute overall hardest false negatives
    all_summary_dfs = []

    # We'll also collect summary rows (one per experiment folder) for a final model-level summary
    final_summary_rows = []

    for experiment_folder in experiment_folders:
        logger.info(f"\n========== Evaluating Experiment: {experiment_folder} ==========")

        # Initialize all models (just architecture definitions)
        models_dict = {}
        for name in model_names:
            try:
                model = initialize_model(model_name=name, num_classes=num_classes, pretrained=False)
                models_dict[name] = model
                logger.info(f'Initialized model: {name}')
            except ValueError as ve:
                logger.error(ve)

        # Evaluate models and obtain false negatives statistics
        evaluation_results, model_false_negatives, dataset_file_paths = evaluate_on_L2(
            models=models_dict,
            transform=transform,
            num_classes=num_classes,
            device=device,
            experiment_folder=experiment_folder
        )

        # If no models were evaluated, skip this experiment
        if not evaluation_results:
            logger.warning(f"No models were successfully evaluated in {experiment_folder}. Skipping.")
            continue

        # Create a results DataFrame (sorted by number of detections in descending order)
        rows = []
        for model_name, scores in evaluation_results.items():
            rows.append({
                'Model': model_name,
                'Recall': scores.get('Recall', 'N/A'),
                'Detections': scores.get('Detections', 'N/A'),
                'FalseNegatives': scores.get('FalseNegatives', 'N/A'),
                'Total_Samples': scores.get('Total', 'N/A')
            })
        results_df = pd.DataFrame(rows)
        results_df = results_df.sort_values(by='Detections', ascending=False).reset_index(drop=True)
        results_df.insert(0, 'Rank', np.arange(1, len(results_df) + 1))

        logger.info("\n=== Evaluation Results on L2 (Ranked by Detections) ===\n")
        print(results_df.to_string(index=False))

        # Save CSV inside the experiment folder for model-level statistics
        results_dir = os.path.join(experiment_folder, 'results')
        os.makedirs(results_dir, exist_ok=True)
        results_csv_path = os.path.join(results_dir, 'evaluation_L2_false_negative_stats.csv')
        results_df.to_csv(results_csv_path, index=False)
        logger.info(f"Saved evaluation results to '{results_csv_path}'")

        # Compute intersection of false negatives across all evaluated models
        evaluated_models = list(model_false_negatives.keys())
        if evaluated_models:
            intersection_false_negatives = set(dataset_file_paths)
            for model in evaluated_models:
                intersection_false_negatives = intersection_false_negatives.intersection(set(model_false_negatives[model]))
            if len(intersection_false_negatives) == 0:
                logger.info("All lens samples were recovered by at least one model (no sample was missed by every model).")
            else:
                logger.info(f"{len(intersection_false_negatives)} lens sample(s) were missed by every evaluated model:")
                for f in intersection_false_negatives:
                    logger.info(f"Missed by all models: {f}")
        else:
            logger.warning("No evaluated models found. Skipping false negative intersection computation.")

        # Build a per-sample summary across models.
        summary_rows = []
        for f in dataset_file_paths:
            missed_by = [model for model in evaluated_models if f in model_false_negatives[model]]
            summary_rows.append({
                'Sample': os.path.basename(f),
                'Missed_By_Count': len(missed_by),
                'Missed_By': ", ".join(missed_by) if missed_by else "Detected by all"
            })
        summary_df = pd.DataFrame(summary_rows)
        # Sort the summary by "Missed_By_Count" from highest to lowest
        summary_df = summary_df.sort_values(by='Missed_By_Count', ascending=False).reset_index(drop=True)
        logger.info("\n=== False Negative Summary Across Models (Sorted by Missed Count) ===\n")
        print(summary_df.to_string(index=False))
        summary_csv_path = os.path.join(results_dir, 'false_negative_summary.csv')
        summary_df.to_csv(summary_csv_path, index=False)
        logger.info(f"Saved false negative summary to '{summary_csv_path}'")

        # Save this experiment's summary for overall aggregation
        all_summary_dfs.append(summary_df)

        # Create a mapping from sample basename to its full file path
        file_path_dict = {os.path.basename(f): f for f in dataset_file_paths}

        # --------------------- Plot Insights Using Matplotlib for This Experiment ---------------------
        # 1. Bar Chart of Top N Hardest Samples (by Missed_By_Count)
        top_N = 10
        top_summary_df = summary_df.head(top_N)
        plt.figure(figsize=(10, 6))
        plt.bar(top_summary_df['Sample'], top_summary_df['Missed_By_Count'])
        plt.xticks(rotation=45, ha='right')
        plt.xlabel('Sample')
        plt.ylabel('Missed By Count')
        plt.title(f'Top {top_N} Hardest Samples in Experiment\n(Missed by Most Models)')
        bar_chart_path = os.path.join(results_dir, 'hardest_samples_bar_chart.png')
        plt.tight_layout()
        plt.savefig(bar_chart_path, bbox_inches='tight')
        plt.show()
        logger.info(f"Saved bar chart of hardest samples to '{bar_chart_path}'")

        # 2. Visualize the Hardest Samples (Display their images)
        top_N_images = 5
        top_images_df = summary_df.head(top_N_images)
        n_images = len(top_images_df)
        plt.figure(figsize=(15, 3))
        for i, row in top_images_df.iterrows():
            sample_basename = row['Sample']
            missed_count = row['Missed_By_Count']
            full_path = file_path_dict.get(sample_basename, None)
            if full_path is None:
                continue
            # Load the image using the helper function
            try:
                image_array = load_and_process_fits_image(full_path)
            except Exception as e:
                logger.error(f"Error processing image {full_path}: {e}")
                continue
            plt.subplot(1, n_images, i + 1)
            plt.imshow(image_array)
            plt.title(f"{sample_basename}\nMissed: {missed_count}")
            plt.axis('off')
        plt.tight_layout()
        images_chart_path = os.path.join(results_dir, 'hardest_samples_images.png')
        plt.savefig(images_chart_path, bbox_inches='tight')
        plt.show()
        logger.info(f"Saved images of hardest samples to '{images_chart_path}'")

        # Collect the best (rank 1) model row for final summary (optional)
        best_row = results_df.iloc[0].to_dict()
        best_row['Experiment'] = os.path.basename(experiment_folder)
        final_summary_rows.append(best_row)

    # ============================ Overall False Negative Summary Across Experiments ============================
    if all_summary_dfs:
        # Concatenate all summaries (they share the same "Sample" names)
        overall_df = pd.concat(all_summary_dfs)
        # Group by "Sample" and sum the "Missed_By_Count" across experiments
        overall_summary = overall_df.groupby("Sample", as_index=False)["Missed_By_Count"].sum()
        overall_summary = overall_summary.sort_values(by="Missed_By_Count", ascending=False).reset_index(drop=True)
        logger.info("\n=== Overall False Negative Summary Across Experiments ===\n")
        print(overall_summary.to_string(index=False))
        
        # Plot overall bar chart for top overall hardest samples
        overall_top_N = 10
        overall_top_summary = overall_summary.head(overall_top_N)
        plt.figure(figsize=(10, 6))
        plt.bar(overall_top_summary['Sample'], overall_top_summary['Missed_By_Count'])
        plt.xticks(rotation=45, ha='right')
        plt.xlabel('Sample')
        plt.ylabel('Total Missed Count (Across Experiments)')
        plt.title(f'Top {overall_top_N} Hardest Samples Overall\n(Aggregated Missed Count)')
        overall_bar_chart_path = os.path.join("overall_results", 'overall_hardest_samples_bar_chart.png')
        os.makedirs("overall_results", exist_ok=True)
        plt.tight_layout()
        plt.savefig(overall_bar_chart_path, bbox_inches='tight')
        plt.show()
        logger.info(f"Saved overall bar chart of hardest samples to '{overall_bar_chart_path}'")
        
        # Visualize overall hardest samples images (top 5)
        overall_top_N_images = 5
        overall_top_images = overall_summary.head(overall_top_N_images)
        plt.figure(figsize=(15, 3))
        # We reuse file_path_dict from the last experiment (dataset is same across experiments)
        for i, row in overall_top_images.iterrows():
            sample_basename = row['Sample']
            missed_total = row['Missed_By_Count']
            full_path = file_path_dict.get(sample_basename, None)
            if full_path is None:
                continue
            try:
                image_array = load_and_process_fits_image(full_path)
            except Exception as e:
                logger.error(f"Error processing image {full_path}: {e}")
                continue
            plt.subplot(1, overall_top_N_images, i + 1)
            plt.imshow(image_array)
            plt.title(f"{sample_basename}\nTotal Missed: {missed_total}")
            plt.axis('off')
        plt.tight_layout()
        overall_images_chart_path = os.path.join("overall_results", 'overall_hardest_samples_images.png')
        plt.savefig(overall_images_chart_path, bbox_inches='tight')
        plt.show()
        logger.info(f"Saved overall images of hardest samples to '{overall_images_chart_path}'")
    else:
        logger.warning("No experiment summaries available for overall aggregation.")

    # ============================ Final Model-Level Summary Across Experiments ============================
    if final_summary_rows:
        final_summary_df = pd.DataFrame(final_summary_rows)
        cols = ['Experiment', 'Rank', 'Model', 'Recall', 'Detections', 'FalseNegatives', 'Total_Samples']
        final_summary_df = final_summary_df[cols]
        logger.info("\n============= Final Model-Level Summary Across Experiments =============\n")
        print(final_summary_df.to_string(index=False))
    else:
        print("No experiments had valid results. Nothing to summarize.")
```
